# Business Relevant Feature Selection

The dataset contains 53 columns. However, not all columns are relevant to the business problem.

Our objective is to identify the features that help explain delivery delays and remove irrelevant information from further analysis.

```python
for i, col in enumerate(data.columns):
    print(f"{i+1}.{col}")
```

```python
clean_data = data.copy()
```

```python
columns_to_drop = [
    'Customer Email',
    'Customer Fname',
    'Customer Lname',
    'Customer Password',
    'Customer Street',
    'Customer Id',
    'Product Image',
    'Order Customer Id',
    'Customer Zipcode',
    'Category Id',
    'Department Id',
    'Product Card Id',
    'Order Item Cardprod Id'
]

clean_data = clean_data.drop(columns=columns_to_drop)
```
